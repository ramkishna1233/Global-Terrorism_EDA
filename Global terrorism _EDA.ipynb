```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',200)
```

```python
#Loading dataset
data=pd.read_csv(r"C:\Users\prama\Desktop\Spark intern files\Global Terrorism - START data\globalterrorismdb_0718dist.csv")
```

```python
#checking the shape of the data
data.shape
```

```python
len(data)
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#Checking the missing value percentage
data.isnull().sum()*100/len(data)
```

```python
#Filtering columns with null values
nullcols=(data.isna().sum()/len(data)*100)
nullcols
```

```python
nullcols=nullcols[nullcols>35]
```

```python
len(nullcols)
```

```python
type(nullcols)
```

```python
nullcols=list(nullcols.index)
```

```python
#Removing columns with null vaues
data=data.drop(labels=nullcols,axis=1)
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.columns
```

```python
#Renaming the column names
data.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'Country','region_txt':'Region','city':'City','latitude':'Latitude',
                    'longitude':'Longitude','targtype1_txt':'Targetype','targsubtype1_txt':'Targetsubtype','weaptype1_txt':'Weapon','gname':'groupName','nkill':'kill',
                    'attacktype1_txt':'attacktype'},inplace=True)
```

```python
#Filtering the dataset by selecting requried columns for our analysis
data=data[['year','month','day','Country','Region','City','Latitude','Longitude','Targetype','Targetsubtype','Weapon','groupName','kill','attacktype']]
```

```python
data.head()
```

```python
#Checking the missing value percentage for the new dataset
data.isna().sum()*100/len(data)
```

```python
#Finding the statastical description of the new dataset
data.describe()
```

```python
data['City'].value_counts()
```

```python
#Imputing missing values.
data['City']=data['City'].fillna('Unknown')
```

```python
data['Targetsubtype'].value_counts()
```

```python
data['Targetsubtype']=data['Targetsubtype'].fillna('Unnamed Civilian/Unspecified')
```

```python
data.isna().sum()
```

```python
data['kill'].value_counts()
```

```python
data['kill']=data['kill'].fillna(0)
```

```python
data['Longitude']=data['Longitude'].ffill()
data['Latitude']=data['Latitude'].ffill()
```

```python
data.isna().sum()
```

```python
#Checking years with most attacks by plotting bar graph
year=data['year'].unique()
year_count=data['year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(15,10))
sns.barplot(x=year,y=year_count)
plt.xticks(rotation=90)
plt.xlabel('Attacking year',fontsize=15)
plt.ylabel('Number of attacks in each year',fontsize=15)
plt.title('Attacks in years',fontsize=25)
plt.show()
```

```python
data['Country'].value_counts()
```

```python
#Plot a graph to check top attacking countries
plt.figure(figsize=(15,10))
sns.barplot(data=data,x=data['Country'].value_counts()[:10].index,y=data['Country'].value_counts()[:10].values)
plt.title('Top  attacked countries',fontsize=25)
plt.xlabel('country',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
data['City'].value_counts()
```

```python
#Plotting a graph to check top attacked cities
plt.figure(figsize=(20,15))
sns.barplot(data=data,x=data['City'].value_counts()[:10].index,y=data['City'].value_counts()[:10].values)
plt.title('Top Attacked Cities',fontsize=25)
plt.xlabel('Cities',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.show()
```

```python
data['Region'].value_counts()
```

```python
plt.figure(figsize=(20,15))
data[['attacktype','kill']].groupby(['attacktype'],axis=0).sum().plot(kind='bar')
plt.title('Attacking types',fontsize=25)
plt.xlabel('Attacktypes',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
#Check region wise attacks
plt.figure(figsize=(20,15))
sns.barplot(data=data,x=data['Region'].value_counts().index,y=data['Region'].value_counts().values)
plt.xticks(rotation=90)
plt.title('Region wise attacks',fontsize=25)
plt.xlabel('Regions',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.show()
```
